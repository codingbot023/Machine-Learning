```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("diamond.csv")
```

```python
data.head()
```

```python
data.duplicated().sum()
```

```python

data=data.drop_duplicates()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
Cut=LabelEncoder()
data["Cut"]=Cut.fit_transform(data["Cut"])
Color=LabelEncoder()
data["Color"]=Color.fit_transform(data["Color"])
Clarity=LabelEncoder()
data["Clarity"]=Clarity.fit_transform(data["Clarity"])
Polish=LabelEncoder()
data["Polish"]=Polish.fit_transform(data["Polish"])
Symmetry=LabelEncoder()
data["Symmetry"]=Symmetry.fit_transform(data["Symmetry"])
Report=LabelEncoder()
data["Report"]=Report.fit_transform(data["Report"])
```

```python
x=data.drop("Price",axis=1)
y=data["Price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
model=AdaBoostRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model1=AdaBoostRegressor(estimator=DecisionTreeRegressor(max_depth=6),learning_rate=0.6)
model1.fit(x_train,y_train)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
model2=GradientBoostingRegressor()
model2.fit(x_train,y_train)
```

```python
model2.score(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
model3 = GradientBoostingRegressor(max_depth=6, learning_rate=0.6)
model3.fit(x_train,y_train)
```

```python
model3.score(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

```python
model4=XGBRegressor()
model4.fit(x_train,y_train)
```

```python
model4.score(x_train,y_train)
```

```python
model4.score(x_test,y_test)
```
