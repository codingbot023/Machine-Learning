```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
data =pd.read_csv("automobile.csv")
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
imputer = SimpleImputer(strategy ="most_frequent")
data["normalized-losses"]=imputer.fit_transform(data[['normalized-losses']])
```

```python
data["num-of-doors"].value_counts()
```

```python
data["num-of-doors"]=data["num-of-doors"].fillna("four")
```

```python
data.isnull().sum()
```

```python
#get all object column names
for col in data:
    if data[col].dtypes=='O':
        print("Column name : ",col)
        print(data[col].unique())
```

### Converting number names to number

```python
# pip install word2number
```

```python
from word2number.w2n import word_to_num
```

```python
data["num-of-doors"] = data["num-of-doors"].apply(word_to_num)
```

```python
data["num-of-cylinders"] = data["num-of-cylinders"].apply(word_to_num)
```

```python
data.head()
```

```python
data.dtypes
```

### Replacing '?'  from columns

```python
data=data.replace('?',0)
```

```python
# astype is used to convert datatypes
ob_columns = ['bore','stroke','horsepower','peak-rpm']
for col in ob_columns:
    data[col] = data[col].astype('float')
    data[col] = data[col].replace(0,data[col].mean())
    print(col, data[col].dtype)
```

```python
data.dtypes
```

### to find remaining object values

```python
for col in data:
    if data[col].dtype=='O':
        print(col)
```

```python
make=LabelEncoder()
data['make']=make.fit_transform(data['make'])

fuel=LabelEncoder()
data['fuel-type']=fuel.fit_transform(data['fuel-type'])

aspiration=LabelEncoder()
data['aspiration']=aspiration.fit_transform(data['aspiration'])

body=LabelEncoder()
data['body-style']=body.fit_transform(data['body-style'])

drive=LabelEncoder()
data['drive-wheels']=drive.fit_transform(data['drive-wheels'])

engine=LabelEncoder()
data['engine-location']=engine.fit_transform(data['engine-location'])

engine=LabelEncoder()
data['engine-type']=engine.fit_transform(data['engine-type'])

fuel=LabelEncoder()
data['fuel-system']=fuel.fit_transform(data['fuel-system'])
```

```python
data.head()
```

```python
x=data.drop('price',axis=1)
y=data['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
model1=RandomForestRegressor(random_state=10)
model1.fit(x_train,y_train)
print("train score: ",model1.score(x_train,y_train))
print("test score: ",model1.score(x_test,y_test))
```

```python
model2 =RandomForestRegressor(n_estimators=50,random_state=11) 
model2.fit(x_train,y_train)
print("train score: ",model2.score(x_train,y_train))
print("test score: ",model2.score(x_test,y_test))
```
