```python
import pandas as pd
from sklearn.svm  import SVC
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
x =data.drop('variety',axis=1)
y =data['variety']
```

#### finding best parameters for a single model

```python
model =SVC()
gsvc =GridSearchCV(model,
                   {'gamma':['auto','scale'],
                   'C':[1,3,5,7,10,15,20],
                   'kernel':['linear','rbf','poly','sigmoid']},cv=5)
```

```python
gsvc.fit(x,y)
```

```python
gsvc.best_params_
```

```python
gsvc.best_score_
```

```python
pd.DataFrame(gsvc.cv_results_)
```

```python
model1 =KNeighborsClassifier()
gsvc1 =GridSearchCV(model1,
                   {'n_neighbors':[1,3,5,7,10,15,20]},
                    cv=5)
```

```python
gsvc1.fit(x,y)
```

```python
gsvc1.best_params_
```

```python
gsvc1.best_score_
```

```python
model2 =RandomForestClassifier()
gsvc2 =GridSearchCV(model2,
                   {'n_estimators':[1,3,5,7,10,15,20]},
                    cv=5)
```

```python
gsvc2.fit(x,y)
```

```python
gsvc2.best_params_
```

```python
gsvc2.best_score_
```

```python
pd.DataFrame(gsvc2.cv_results_)
```

```python
pd.DataFrame({"Model":["SVC","RandomForest","KNN"],"Best_param":[gsvc.best_params_,gsvc1.best_params_,gsvc2.best_params_],
             "Best_score":[gsvc.best_score_,gsvc1.best_score_,gsvc2.best_score_]})
```
