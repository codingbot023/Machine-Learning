```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
data=pd.read_csv("Country-data.csv")
data
```

```python
new_data=data.drop("country",axis=1)
new_data
```

```python
linkage_data=linkage(new_data,method="ward")#linkage is finding similarity between points,ward calculates the varience between points and choose the points with less varience to cluster
dendro=dendrogram(linkage_data)
```

```python
dendro=dendrogram(linkage_data)
plt.axhline(y=200000,color="red",linestyle="--")#to draw horizontal line with y value 200000
```

```python
hierarchical_cluster=AgglomerativeClustering(n_clusters=2,linkage="ward")
clusters=hierarchical_cluster.fit_predict(new_data)
clusters
```

```python
data["clusters"]=clusters
```
